### Nhiệm vụ 1: Import các thư viện cần thiết và đọc dataset vào dataframe


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

** Đọc file cvs vào dataframe df **

```python
df = pd.read_csv('supermarket_sales.csv')
df.set_index('Invoice ID', inplace=True)
df.head()
```

** Kiểm tra infor() của df **

```python
df.info()
```

```python
df.describe()
```

### Nhiệm vụ 2: Tiền xử lý dữ liệu

*Nhiệm vụ 2.1: Thêm cột Month và Hour*

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'])
df['Month']= df['Date'].dt.month
df['Hour'] = df['Time'].dt.hour
df.head(3)
```

*Nhiệm vụ 2.2 Thêm cột Part_Of_The_Day thể hiện các buổi trong ngày*

```python
time = []
for i in df['Hour']:
    if i>9 and i < 13 :
        time.append('Morning')
    elif i> 12 and i < 18:
        time.append('Afternoon') 
    else:
        time.append('Evening')
df['Part_Of_The_Day'] = time
```

### Nhiệm vụ 3: Báo cáo

*Nhiệm vụ 3.1: Tháng có lợi nhuận cao nhất, lợi nhuận tháng đó là bao nhiêu*

```python
sale_of_moth = df.groupby(by='Month').agg({'gross income':np.sum}).reset_index()
bars = sns.barplot(data=sale_of_moth,x='Month',y='gross income')
for p in bars.patches:
    bars.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=11, xytext=(0, 10),
                textcoords='offset points')
plt.title('Tổng thu nhập từng tháng quí 1')
```

*Nhiệm vụ 3.2: Buổi nào trong ngày có lợi nhuận trung bình cao nhất tại các thành phố*

```python
city_day = df.groupby(['City' , 'Part_Of_The_Day'])['gross income'].mean().reset_index(drop =False)
city_day
```

```python
plt.figure(figsize = (8,4))
sns.lineplot(x = 'Part_Of_The_Day' , y ='gross income' , hue = 'City' , data = city_day,marker ='X')

plt.title('Lợi nhuận trung bình các buổi trong ngày ở từng thành phố')
```

*Nhiệm vụ 3.3: Có mối tương quan nào giữa giá sản phẩm và số lượng mua của người dùng cho từng dòng sản phẩm không*

```python
data = df[['Product line','Unit price','Quantity']]
sns.relplot(data=data,x='Quantity',y='Unit price')
```

*Nhiệm vụ 3.4: Có mối tương quan nào giữa lợi nhuận cửa hàng với đánh giá của khách hàng không*

```python
sns.scatterplot(x=df['Rating'],y=df['gross income'])
```

*Nhiệm vụ 3.5: Giữa khách hàng bình thường và khách hàng thành viên, nhóm nào chi tiêu tại siêu thị nhiều hơn?*

```python
normal_members = df.groupby('Customer type').agg({'gross income': np.sum})
bars = sns.barplot(x=normal_members.index, y='gross income', data=normal_members)
for p in bars.patches:
    bars.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=11, xytext=(0, 10),
                textcoords='offset points')
plt.title('Mức chi tiêu giữa các nhóm khách hàng')
```

*Nhiệm vụ 3.6: Dòng sản phẩm phổ biến cho mỗi loại khách hàng là gì*

```python
plt.figure(figsize = (12,4))

bar = sns.countplot(x = 'Product line' , hue= 'Customer type' , data =df )

plt.legend(loc="upper left")

bar.bar_label(bar.containers[0])
bar.bar_label(bar.containers[1])

plt.title('Dòng sản phẩm phổ biến tương ứng với từng loại khách hàng')
```

*Nhiệm vụ 3.7: Nên tập trung phát triển các sản phẩm trọng tâm nào tại từng thành phố để tối đa hóa lợi nhuận*

```python
product_city =  df.groupby(['City' , 'Product line'])['Total'].sum().reset_index()

plt.figure(figsize = (12,4))

ap =sns.barplot(x =  'Product line', y = 'Total' , hue = 'City' ,data = product_city)

#in dữ liệu lên các cột
ap.bar_label(ap.containers[0] , fontsize= 9)
ap.bar_label(ap.containers[1] , fontsize = 9)
ap.bar_label(ap.containers[2] , fontsize = 9)

plt.legend(loc= 'upper left' , prop ={'size' : 7})
plt.title('Tổng chi tiêu cho từng loại sản phẩm ở từng thành phố')
```
